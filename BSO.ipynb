```python
def ord(itens):
    random.shuffle(itens)
    return itens[:]
```

```python
def ini_comeia(populacao_size, itens): # inincializar uma comeia
    comeia = [] # comeia
    for _ in range(populacao_size): # andar por toda a comeia
        solucao = ord(itens) # crial uma solucao aleatoria para cada abelha
        velocidade = [0] * len(solucao) # inicia a velocidade de cada abelha com 0
        comeia.append({'solucao': solucao, 'velocidade': velocidade, 'mel_local': solucao[:]}) # adiciona a abelha a comeia
    return comeia # retorna a comeia

solucao = ini_comeia(populacao_size, itens)
print(f"caixas: {len(solucao)}")
for i, caixa in enumerate(solucao):
    print(f"Bin {i + 1}: {caixa['solucao']}")
```

```python
def caixas_usadas(itens, peso_caixa):
    caixas = [] # itens dentro da minha caixa
    for item in itens: # andar por todos os itens
        check = False # saber se o item foi ou não colocado em uma caixa ja existente
        for caixa in caixas: # andar por cada caixa
            soma_itens = sum(caixa) # soma o peso de cada item 
            if soma_itens + item <= peso_caixa: # posso colocar esse novo item dentro da minha caixa
                caixa.append(item) # colocando esse item na minha caixa
                check = True # item foi colocado
                break
        if check==False: # criar uma nova caixa para colocar o item
            caixas.append([item])
    return caixas, len(caixas) # retorno as coixas que eu montei com os seu pacotes
```

```python
def BSO(itens, MAX_interacao):
    comeia = ini_comeia(populacao_size, itens) # inicializa a comeia
    melhor_global = {'solucao': comeia[0]['solucao'], 'caixas_usadas': caixas_usadas(comeia[0]['solucao'], peso_caixa)[1]} # melhor solucao global
    for it in range(MAX_interacao): # fazer todas as interacoes
        for abelha in comeia: # andar por cada abelha
            _, solucao_atual = caixas_usadas(abelha['solucao'], peso_caixa) # melhor solucao atual da abelha atual
            melhor_local = caixas_usadas(abelha['mel_local'], peso_caixa)[1] # melhor local da abelha atual
            
            if solucao_atual < melhor_local: # se solucao atual for melhor que solucao local
                abelha['mel_local'] = abelha['solucao'][:] # atualizar o melhor local

            if solucao_atual < melhor_global['caixas_usadas']: # se solucao atual for melhor que solucao global
                melhor_global['solucao'] = abelha['solucao'][:]
                melhor_global['caixas_usadas'] = solucao_atual

            nova_velocidade = []
            nova_solu = []
            for i in range(len(abelha['solucao'])): # passar por cada iten de uma abelha
                #calculo velocidade
                inercia = w * abelha['velocidade'][i] # inercia
                memoria_global = fb1 * random.random() * (melhor_global['solucao'][i] - abelha['solucao'][i]) # influencia pelo melhor global
                memoria_local = fb2 * random.random() * (abelha['mel_local'][i] - abelha['solucao'][i]) # influencia pelo melhor local
                nova_velocidade.append(inercia + memoria_global + memoria_local) # calculo da velocidade
            abelha['velocidade'] = nova_velocidade # atualiza a velocidade da abelha
            
            # calculo para nova posicao
            for i in range(len(abelha['solucao'])):
                nova_pos = abelha['solucao'][i] + int(abelha['velocidade'][i]) # Atualiza a posição com base na nova velocidade
                # uso do raio de busca para limitar as mudancas nas solucoes
                if abs(nova_pos - abelha['solucao'][i] > r):
                    nova_pos = abelha['solucao'][i] + r if nova_pos > abelha['solucao'][i] else abelha['solucao'][i] - r
                nova_solu.append(nova_pos)
            abelha['solucao'] = nova_solu # atualiza a solucao da abelha

        # Piores abelhas
        Piores_abelhas = sorted(comeia, key=lambda b: caixas_usadas(b['solucao'], peso_caixa)[0], reverse=True)[:sb] # ordena as abelhar por ordem da com pior caixas_usadas
        for abelha in Piores_abelhas: # ando por cada abelha
            abelha['solucao'] = ord(itens) # crial uma solucao aleatoria para cada abelha
            abelha['velocidade'] = [0] * len(abelha['solucao']) # Redefinir a velocidade para zero

        print(f"It {it+1}/{MAX_interacao} - Numero de caixas: {melhor_global['caixas_usadas']}")

    return melhor_global
```
